### ***CODIGO GENERADO POR GIT HUB***

```python
# metodo de simpson
import math
import numpy as np
import matplotlib.pyplot as plt

# input's
# n=int(input("ingrese el numero de intervalos: "))   # numero de intervalos
# a=int(input("ingrese el limite inferior: "))        # limite inferior
# b=int(input("ingrese el limite superior: "))        # limite superior
# f=input("ingrese la funcion: ")                     # funcion a integrar
# ingreso de datos
n= 30                         # numero de intervalos
a= 0                         # limite inferior
b= 4                          # limite superior
f = 'np.sin(x**2)'            # funcion a integrar
# metodo de simpson
h=(b-a)/n                                           # ancho de los intervalos
x=np.arange(a,b+h,h)                                # vector de nodos
y=eval(f)                                           # evaluar la funcion en los nodos
s=y[0]+y[n]                                         # sumar los nodos 0 y n

for i in range(1,n):                                # sumar los nodos intermedios
    if i%2==0:                                      # si es par
        s+=2*y[i]
    else:                                           # si es impar
        s+=4*y[i]
s=s*h/3                                             # multiplicar por h/3
print("el resultado de la integral es: ",s)         # imprimir resultado
# graficar
plt.plot(x,y)
plt.fill_between(x,y)
plt.title("Metodo de Simpson")
plt.show()
```

### ***CODIGO DE CHAT GPT***

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a integrar
def funcion_a_integrar(x):
    return (x**3+x**2-84156*x)/(1+x)**0.5 # Función a integrar

# Definir el método de Simpson
def metodo_simpson(a, b, n):
    """
    Aproxima la integral definida de la función en el intervalo [a, b]
    utilizando el método de Simpson con n subintervalos.
    """
    h = (b - a) / n

    suma = funcion_a_integrar(a) + funcion_a_integrar(b)

    for i in range(1, n, 2):
        suma += 4 * funcion_a_integrar(a + i * h) # Sumar las evaluaciones en posiciones impares

    for i in range(2, n-1, 2):
        suma += 2 * funcion_a_integrar(a + i * h) # Sumar las evaluaciones en posiciones pares

    integral_aproximada = h / 3 * suma
    return integral_aproximada

# Definir el intervalo [a, b] y el número de subintervalos n
a = 1
b = 500
n = 8  # Número de subintervalos (debe ser par para el método de Simpson)

# Calcular la integral aproximada usando el método de Simpson
resultado_aproximado = metodo_simpson(a, b, n)
print(f'El resultado de la integral es aproximadamente {resultado_aproximado:.4f}')

# Crear datos para el gráfico
x_vals = np.linspace(a, b, 100)
y_vals = funcion_a_integrar(x_vals)

# Graficar la función y resaltar el área bajo la curva
plt.plot(x_vals, y_vals, label='Función a integrar')
plt.fill_between(x_vals, y_vals, where=[(a <= x <= b) for x in x_vals], alpha=0.3, label='Área aproximada')
plt.title('Método de Simpson para la integración numérica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Mostrar el resultado de la aproximación en la esquina superior izquierda
plt.annotate(f'Aproximación: {resultado_aproximado:.4f}', xy=(0.05, 0.9), xycoords='axes fraction', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", edgecolor="none", facecolor="white"))

# Mostrar el gráfico
plt.show()
```
