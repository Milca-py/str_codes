```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# FUNCTION TO CALCULATE THE EUCLIDEAN DISTANCE BETWEEN TWO POINTS
def radio(z, Io):
    r = (z**(6/5)/(2*np.pi*Io/3)**(2/5)-z**2)**0.5
    return r


def influencia(z, r):
    Io = (1/z**2)*(3/(2*np.pi))*(1/(1+(r/z)**2)**(5/2))
    return Io
```

```python
# CALCULOS
Io = np.arange(1, 11, 1)


# creacion de vulvo de presiones para una carga puntual

φ = np.arange(0, 95, 5)

Sφ = np.sin(np.deg2rad(φ))

zf = np.sqrt(3/(2*np.pi*Io))


Z = []

for i in zf:

    Z.append(i*Sφ)

r = []

for i in range(len(Io)):

    r.append(radio(Z[i], Io[i]))

r = np.nan_to_num(r, nan=0)

negr = []

for i in range(len(Io)):

    negr.append(-r[i])
```

```python
# PLOTEO DE LOS RESULTADOS
color = []
for i in range(len(Io)):
    color.append(plt.cm.jet(i/len(Io)))

plt.figure(figsize=(10, 10))
for i in range(len(Io)):
    plt.plot(r[i], Z[i], color=color[i])
    plt.plot(negr[i], Z[i], color=color[i])

plt.axhline(0, color='black', linestyle='--')
plt.gca().invert_yaxis()
plt.xlabel('R')
plt.ylabel('Z')
plt.title('Gráfico de Z vs. R para diferentes valores de Io')
plt.show()
```
